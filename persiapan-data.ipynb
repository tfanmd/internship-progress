**Penyaringan Emas Murni WikiMatrix**

```python
import pandas as pd
import string
import re

print("Langkah 15: Meluncurkan Mesin Penyaring Emas untuk WikiMatrix...")

# Fungsi pembantu untuk membongkar kembar gaib (spasi tak terlihat, beda kapital, dll)
def normalisasi_total(teks):
    if not isinstance(teks, str):
        return ""
    teks = teks.lower().strip()
    # Hapus semua tanda baca
    teks = teks.translate(str.maketrans('', '', string.punctuation))
    # Hapus semua spasi agar menjadi satu kata dempet
    return "".join(teks.split())

# Membuat copy data dari hasil Langkah 14 kemarin
df_saring = df_wikimatrix.copy()

# LAPIS 1: Saring Kembar Gaib via Normalisasi
df_saring['indo_norm'] = df_saring['indonesia'].apply(normalisasi_total)
df_saring['jawa_norm'] = df_saring['jawa'].apply(normalisasi_total)
df_saring = df_saring[df_saring['indo_norm'] != df_saring['jawa_norm']]

# LAPIS 2: Wajib mengandung kata kunci Bahasa Jawa asli
kata_kunci_jawa = [
    'sing', 'ing', 'lan', 'saka', 'saking', 'wonten', 'ora', 'mboten', 
    'yaiku', 'iki', 'kuwi', 'dipun', 'piyambak', 'nanging', 'kagem', 'ipun',
    'utawa', 'marang', 'menika', 'sawise', 'amarga', 'mung', 'kalebu', 'iku'
]
# Membuat regex pattern kata utuh (\b memastikan kata berdiri sendiri, bukan bagian dari kata lain)
pattern_jawa = '|'.join([r'\b' + kata + r'\b' for kata in kata_kunci_jawa])
df_saring = df_saring[df_saring['jawa'].str.lower().str.contains(pattern_jawa, regex=True, na=False)]

# LAPIS 3: Blacklist kata hubung Indonesia yang sering mengontaminasi kolom Jawa
kata_larangan_indo = ['adalah', 'yang', 'dengan', 'bahwa']
pattern_indo = '|'.join([r'\b' + kata + r'\b' for kata in kata_larangan_indo])
df_saring = df_saring[~df_saring['jawa'].str.lower().str.contains(pattern_indo, regex=True, na=False)]

# Membersihkan kolom bantu normalisasi dan mereset index
df_wikimatrix_clean = df_saring.drop(columns=['indo_norm', 'jawa_norm']).reset_index(drop=True)

print(f"📊 Dari 19.595 data WikiMatrix yang kotor...")
print(f"✅ Berhasil disaring ketat menjadi: {len(df_wikimatrix_clean)} baris EMAS MURNI!")

# Menampilkan 10 data teratas untuk pembuktian kualitas
display(df_wikimatrix_clean.head(10))
```

**Membersihkan Simbol **

```python
print("Mulai membersihkan simbol Mojibake (Î± -> α) pada dataset final...")

# 1. Load dataset final yang kemarin sudah jadi
df_fix = pd.read_csv("dataset_final_20k.csv")

# 2. Fungsi sakti untuk mengembalikan kacamata baca ke UTF-8
def perbaiki_mojibake(teks):
    if not isinstance(teks, str):
        return teks
    try:
        # Mengubah teks rusak Latin-1 kembali ke biner asli, lalu didecode ulang ke UTF-8
        return teks.encode('latin1').decode('utf-8')
    except:
        return teks

# 3. Terapkan fungsi ke kolom indonesia dan jawa
df_fix['indonesia'] = df_fix['indonesia'].apply(perbaiki_mojibake)
df_fix['jawa'] = df_fix['jawa'].apply(perbaiki_mojibake)

# 4. Overwrite (timpa) kembali ke file CSV yang sama
df_fix.to_csv("dataset_final_20k.csv", index=False)

print("✅ Selesai! Simbol rusak berhasil disembuhkan.")
print("Contoh teks yang tadinya rusak sekarang menjadi:")
# Menampilkan baris yang mengandung kata Centauri buat pembuktian
display(df_fix[df_fix['indonesia'].str.contains('Centauri', na=False, case=False)].head(1))
```

**Cek baris yang terbalik**

```python
import re

print("Langkah Diagnostik 1: Mendeteksi Kolom yang Terbalik...")

# PENTING: Pastikan nama 'df_final' di bawah ini diganti sesuai dengan
# nama variabel DataFrame gabungan 20.000 baris yang ada di notebook lu sekarang.
df_audit = df_final.copy()

# Daftar kata kunci Bahasa Jawa unik (tidak ada di Bahasa Indonesia) sebagai sensor
kata_kunci_jawa = [
    'sing', 'ing', 'lan', 'saka', 'saking', 'wonten', 'ora', 'mboten', 
    'yaiku', 'iki', 'kuwi', 'dipun', 'piyambak', 'nanging', 'kagem', 'ipun',
    'utawa', 'marang', 'menika', 'sawise', 'amarga', 'mung', 'kalebu', 'iku'
]

# Membuat regex pattern untuk mencetak batas kata utuh (\b) agar lebih akurat
pattern_jawa = '|'.join([r'\b' + kata + r'\b' for kata in kata_kunci_jawa])

def hitung_kata_jawa(teks):
    if not isinstance(teks, str):
        return 0
    # Menghitung berapa banyak kata Jawa yang muncul (case-insensitive)
    matches = re.findall(pattern_jawa, teks.lower())
    return len(matches)

# Mulai menghitung bobot bahasa Jawa di kedua kolom
df_audit['skor_jawa_di_indo'] = df_audit['indonesia'].apply(hitung_kata_jawa)
df_audit['skor_jawa_di_jawa'] = df_audit['jawa'].apply(hitung_kata_jawa)

# Logika Terbalik: Jika komponen kata Jawa di kolom Indo lebih dominan
kondisi_terbalik = (df_audit['skor_jawa_di_indo'] > df_audit['skor_jawa_di_jawa']) & (df_audit['skor_jawa_di_indo'] > 0)
df_terbalik = df_audit[kondisi_terbalik]

print(f"\n📊 HASIL AUDIT KOLOM BAHASA:")
print(f"- Total seluruh data yang diperiksa : {len(df_audit)} baris")
print(f"- Terdeteksi POSISI TERBALIK         : {len(df_terbalik)} baris")

if len(df_terbalik) > 0:
    print("\n👀 Sampel 5 data yang terdeteksi TERBALIK oleh sistem:")
    display(df_terbalik[['indonesia', 'jawa']].head(5))
else:
    print("\n✅ Aman! Tidak ditemukan kolom yang terbalik pada sampel data ini.")
```

**Cek Simbol**

```python
print("Langkah Diagnostik 2 (Revisi): Meluncurkan Detektor Karakter Agresif...")

# PENTING: Pastikan variabel 'df_final' sesuai dengan nama DataFrame gabungan 20.000 baris lu
df_audit2_rev = df_final.copy()

# POLA SAKTI (Whitelist Negasi):
# Mencari semua karakter yang BUKAN huruf (a-z, A-Z), BUKAN angka (0-9), 
# BUKAN spasi (\s), BUKAN tanda baca standar, dan BUKAN huruf aksen Jawa (é, è, É, È)
pattern_bukan_normal = r'[^a-zA-Z0-9\s.,\-?!()"\':;_\+=\[\]\{\}/\\&%\*@#<>éèÉÈ]'

def cek_karakter_aneh(teks):
    if not isinstance(teks, str):
        return False
    # Mengembalikan True jika ditemukan karakter liar di luar zona aman
    return bool(re.search(pattern_bukan_normal, teks))

# Memeriksa kolom indonesia dan jawa
df_audit2_rev['kotor_indo'] = df_audit2_rev['indonesia'].apply(cek_karakter_aneh)
df_audit2_rev['kotor_jawa'] = df_audit2_rev['jawa'].apply(cek_karakter_aneh)

kondisi_kotor_agresif = df_audit2_rev['kotor_indo'] | df_audit2_rev['kotor_jawa']
df_kotor_agresif = df_audit2_rev[kondisi_kotor_agresif]

print(f"\n📊 HASIL AUDIT KARAKTER (METODE AGRESIF):")
print(f"- Total seluruh data yang diperiksa       : {len(df_audit2_rev)} baris")
print(f"- Terdeteksi mengandung KARAKTER REWEL/RUSAK: {len(df_kotor_agresif)} baris")

if len(df_kotor_agresif) > 0:
    print("\n👀 Sampel 10 data yang berhasil dibongkar oleh Detektor Agresif:")
    display(df_kotor_agresif[['indonesia', 'jawa']].head(10))
else:
    print("\n✅ Bersih Sempurna! Tidak ada karakter aneh yang lolos dari sensor.")
```

**Memperbaiki Permasalahan Kolom Terbalik**

```python
print("Memulai Proses Pemulihan Kolom yang Terbalik...")

# Pastikan nama 'df_final' sesuai dengan nama DataFrame 20.074 baris lu saat ini
df_clean_kolom = df_final.copy()

# Kamus Mini Kata Kunci Jawa sebagai kompas penentu arah bahasa
kata_kunci_jawa = [
    'sing', 'ing', 'lan', 'saka', 'saking', 'wonten', 'ora', 'mboten', 
    'yaiku', 'iki', 'kuwi', 'dipun', 'piyambak', 'nanging', 'kagem', 'ipun',
    'utawa', 'marang', 'menika', 'sawise', 'amarga', 'mung', 'kalebu', 'iku'
]
pattern_jawa = '|'.join([r'\b' + kata + r'\b' for kata in kata_kunci_jawa])

def hitung_kata_jawa(teks):
    if not isinstance(teks, str):
        return 0
    return len(re.findall(pattern_jawa, teks.lower()))

# 1. Hitung distribusi bobot kata Jawa di kedua kolom
skor_indo = df_clean_kolom['indonesia'].apply(hitung_kata_jawa)
skor_jawa = df_clean_kolom['jawa'].apply(hitung_kata_jawa)

# 2. Kunci indeks baris yang posisinya terbalik
kondisi_terbalik = (skor_indo > skor_jawa) & (skor_indo > 0)
total_terbalik = kondisi_terbalik.sum()

# 3. EKSEKUSI SWAP: Tukar posisi konten kolom Indonesia dan Jawa secara simultan
# Penggunaan .values di akhir wajib dipasang agar Pandas tidak bingung menyinkronkan indeks saat ditukar
df_clean_kolom.loc[kondisi_terbalik, ['indonesia', 'jawa']] = df_clean_kolom.loc[kondisi_terbalik, ['jawa', 'indonesia']].values

print(f"\n⚡ ELEKTRO-SHOCK SELESAI!")
print(f"- Total baris yang berhasil diluruskan kembali: {total_terbalik} baris.")

# 4. SANITY CHECK (Validasi Ulang Hasil Pemulihan)
skor_indo_baru = df_clean_kolom['indonesia'].apply(hitung_kata_jawa)
skor_jawa_baru = df_clean_kolom['jawa'].apply(hitung_kata_jawa)
sisa_terbalik = ((skor_indo_baru > skor_jawa_baru) & (skor_indo_baru > 0)).sum()

print(f"- Sisa baris yang terbalik setelah diperbaiki  : {sisa_terbalik} baris.")

# 5. Update data utama lu dengan data yang kolomnya sudah lurus
df_final = df_clean_kolom.copy()
print("\n💾 Variabel 'df_final' resmi diperbarui dan siap ke tahap pembersihan simbol!")
```

**Memperbaiki Permasalahan Simbol**

```python
# Membuat salinan dari df_final yang kolomnya sudah lurus
df_clean_simbol = df_final.copy()

def bersihkan_mojibake_dan_simbol(teks):
    if not isinstance(teks, str):
        return teks
    
    # 1. Memperbaiki encoding huruf beraksen (contoh: Ã© kembali menjadi é)
    try:
        teks = teks.encode('latin1').decode('utf-8')
    except:
        pass
        
    # 2. Menghapus karakter Â akibat kesalahan pembacaan spasi khusus
    teks = teks.replace('Â', '')
    
    # 3. Menghapus karakter di luar whitelist (huruf, angka, tanda baca standar, dan aksen é/è)
    pattern_bukan_normal = r'[^a-zA-Z0-9\s.,\-?!()"\':;_\+=\[\]\{\}/\\&%\*@#<>éèÉÈ]'
    teks = re.sub(pattern_bukan_normal, '', teks)
    
    # 4. Mengubah spasi ganda atau berlebih menjadi spasi tunggal
    teks = re.sub(r'\s+', ' ', teks).strip()
    
    return teks

# Menerapkan fungsi pembersihan pada kedua kolom bahasa
df_clean_simbol['indonesia'] = df_clean_simbol['indonesia'].apply(bersihkan_mojibake_dan_simbol)
df_clean_simbol['jawa'] = df_clean_simbol['jawa'].apply(bersihkan_mojibake_dan_simbol)

# Validasi ulang menggunakan standar detektor agresif
def cek_karakter_aneh(teks):
    if not isinstance(teks, str):
        return False
    return bool(re.search(pattern_bukan_normal, teks))

kotor_indo = df_clean_simbol['indonesia'].apply(cek_karakter_aneh)
kotor_jawa = df_clean_simbol['jawa'].apply(cek_karakter_aneh)
sisa_kotor = (kotor_indo | kotor_jawa).sum()

print("Hasil Pembersihan Karakter:")
print(f"- Sisa baris yang mengandung karakter rusak: {sisa_kotor} baris")

# Memperbarui variabel utama jika sisa kotor sudah berkurang/nol
df_final = df_clean_simbol.copy()
```

```python
df = pd.read_csv("dataset_final_bersih_20k.csv")

# 2. Fungsi untuk menghitung jumlah kata
def hitung_kata(teks):
    if not isinstance(teks, str):
        return 0
    return len(teks.split())

# 3. Menghitung jumlah kata di masing-masing kolom
df['len_indo'] = df['indonesia'].apply(hitung_kata)
df['len_jawa'] = df['jawa'].apply(hitung_kata)

# 4. Menghitung rasio panjang kata (perbandingan nilai maksimum dan minimum)
df['rasio_panjang'] = df.apply(
    lambda r: max(r['len_indo'], r['len_jawa']) / max(1, min(r['len_indo'], r['len_jawa'])), 
    axis=1
)

# 5. Menentukan ambang batas anomali (rasio panjang kata > 3.0 atau ada kolom kosong)
threshold = 3.0
kondisi_anomali = (df['rasio_panjang'] > threshold) | (df['len_indo'] == 0) | (df['len_jawa'] == 0)
df_anomali = df[kondisi_anomali]

print("Hasil Audit Rasio Panjang Kalimat:")
print(f"- Total seluruh baris diperiksa : {len(df)} baris")
print(f"- Terdeteksi rasio tidak wajar (> {threshold}x lipat): {len(df_anomali)} baris")

# 6. Menampilkan sampel jika ditemukan anomali panjang kalimat
if len(df_anomali) > 0:
    print("\nSampel data dengan perbedaan panjang ekstrem:")
    pd.set_option("display.max_colwidth", None)
    display(df_anomali[['indonesia', 'len_indo', 'jawa', 'len_jawa']].head(8))
```

```python
df_audit = pd.read_csv("dataset_fixx.csv")

def ekstrak_entitas_kapital(teks):
    if not isinstance(teks, str):
        return set()
    # Mengambil kata yang diawali huruf kapital dengan panjang minimal 2 karakter
    return set(re.findall(r'\b[A-Z][a-zA-Z0-9_]+\b', teks))

# 2. Menerapkan ekstraksi entitas pada kedua kolom
entitas_indo = df_audit['indonesia'].apply(ekstrak_entitas_kapital)
entitas_jawa = df_audit['jawa'].apply(ekstrak_entitas_kapital)

# 3. Logika deteksi: memiliki entitas di kedua kolom, tetapi tidak ada kecocokan sama sekali
kondisi_miskorelasi = []
for idx in range(len(df_audit)):
    set_i = entitas_indo.iloc[idx]
    set_j = entitas_jawa.iloc[idx]
    
    # Jika kedua kolom punya kata kapital, tapi tidak ada yang sama (irisan kosong)
    if len(set_i) > 0 and len(set_j) > 0 and len(set_i.intersection(set_j)) == 0:
        kondisi_miskorelasi.append(True)
    else:
        kondisi_miskorelasi.append(False)

df_miskorelasi = df_audit[kondisi_miskorelasi]

# 4. Menampilkan hasil statistik audit
print("Hasil Audit Miskorelasi Entitas:")
print(f"- Total seluruh baris diperiksa : {len(df_audit)} baris")
print(f"- Terdeteksi tidak sesuai makna : {len(df_miskorelasi)} baris")

# 5. Menampilkan sampel data yang tidak sesuai makna
if len(df_miskorelasi) > 0:
    print("\nSampel data yang terdeteksi miskorelasi:")
    pd.set_option("display.max_colwidth", None)
    display(df_miskorelasi[['indonesia', 'jawa']].head(5))
```
